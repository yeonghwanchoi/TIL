<h4>25. 다음 (student) 클래스에서 자신의 이름, 나이, 성별을 출력하는 who() 함수를 추가하세요.</h4>


```python
class student:
    def __init__(self, name, age, sex):
        self.name = name
        self.age = age
        self.sex = sex

    def who(self):
        print("이름: {} 나이: {} 성별: {}".format(self.name, self.age, self.sex))

itsme = student("최영환", 20, "남자")
itsme.who()    
#클래스 매니저 이름은 예시일뿐입니다
```
