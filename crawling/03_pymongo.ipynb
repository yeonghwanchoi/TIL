```python
# 패키지 불러오기
import pymongo
import pandas as pd

import zigbang as zb # zigbang.py가 본 주피터 노트북 파일과 동일한 경로에 있어야 함
```

```python
# 데이터베이스 서버 연결 - 여러분의 AWS Public IP를 입력해주세요.
client = pymongo.MongoClient("mongodb://your_public_ip_address/:27017/")
```

```python
client
```

```python
# 데이터 베이스와 컬렉션 생성
zigbang = client.crawling.zigbang
zigbang
```

```python
items = zb.oneroom('성수동')
len(items)
```

```python
items
```

```python
# 데이터프레임 저장
sungsoo_df = pd.DataFrame(items)
# 필요한 컬럼만 추출
columns = ["address1", "size_m2", "floor", "sales_type", "rent", "deposit"]
sungsoo_df = sungsoo_df[columns]
sungsoo_df.tail(2)
```

```python
zigbang2 = client.crawling.zigbang2
```

```python
# 딕셔너리 형태로 변환
items = sungsoo_df.to_dict("records")
items[0]
```

```python
ids = zigbang2.insert(items)
```

```python
results = zigbang2.find()
```

```python
QUERY = {"rent": {"$lte": 50}}
results = zigbang2.find(QUERY)
pd.DataFrame(results).tail()
```
