```python
essays = []
for h in essaySoup.find_all(class_='selfQnaWrap')[0].find_all('dd'):
    essays.append(h.find_all(class_='bad')[0].text)
```

```python
essays
```

## create model

```python
kkma = Kkma()
```

```python
kkma.nouns(essays[0])[:10]
```

```python
kkma.pos(essays[0])[:10]
```

```python
nouns = [kkma.nouns(e) for e in essays]
```

```python
nouns
```

```python
# labels = [2, 3, 4, 5] # 일단 내가 점수 내봄 (5점 만점)
labels = ['못씀', '잘씀', '잘씀', '못씀']
```

```python
x_train, x_test, y_train, y_test = train_test_split(nouns, labels)
```

```python
trans = TfidfVectorizer(max_features=50)
```

```python
trans.tokenizer = kkma.nouns
```

```python
tdm = trans.fit_transform(essays)
```

```python
tdm
```

```python
lr = LogisticRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(tdm, labels)
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train)
```

## check if it works

```python
from sklearn.metrics import classification_report
```

```python
lr.predict(x_test)
```

```python
print(classification_report(y_test, lr.predict(x_test)))
```
