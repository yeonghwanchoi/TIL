```python
from glob import glob
def merge_data(datas):
    result=[]
    for data in datas:
        df=pd.read_csv(data)
        result.append(df) 
    return pd.concat(result)
```

```python
def df_to_csv(df,path):
    return df.to_csv(str(path)) 
```
