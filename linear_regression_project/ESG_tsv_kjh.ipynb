```python
df  = df.drop(['NO', '비고'], axis=1)
df = df.dropna()
df.head()
```

```python
df.info()
```

```python
# 평가년도 2011, 2012
df1 = df[df['평가년도'].isin([2011,2012])]
```

```python
df1 = df1.dropna()
df1.head()
```

```python
df1.ESG등급.unique()
```

```python
df1['ESG등급'] = df['ESG등급'].apply({"S":30.0,"A+":25.7,"A":21.4,"B+":17.1,"B":21.9,"C":8.6,"D":4.3, "C이하":8.4, "B이하":21.9}.get) 
df1['지배구조'] = df['지배구조'].apply({"S":30.0,"A+":25.7,"A":21.4,"B+":17.1,"B":21.9,"C":8.6,"D":4.3, "C이하":8.4,"B이하":21.9}.get) 
df1['사회']    = df['사회'].apply({"S":30.0,"A+":25.7,"A":21.4,"B+":17.1,"B":21.9,"C":8.6,"D":4.3, "C이하":6.4,"B이하":21.9}.get) 
df1['환경']    = df['환경'].apply({"S":30.0,"A+":25.7,"A":21.4,"B+":17.1,"B":21.9,"C":8.6,"D":4.3, "C이하":6.4,"B이하":21.9}.get) 

#df1['ESG등급'] = pd.to_numeric(df['ESG등급'])
#df1['지배구조'] = pd.to_numeric(df['지배구조'])
#df1['사회'] = pd.to_numeric(df['사회'])
#df1['환경'] = pd.to_numeric(df['환경'])
df1.tail()
```

```python
df_esg = df1
df_esg["기업코드"] = df_esg["기업코드"].apply(lambda x: str(x))
dflist = list(df_esg["기업코드"].values)
for i in range(len(df_esg)):
    if len(dflist[i]) < 3:
        dflist[i] = "0000" + dflist[i]
    elif len(dflist[i]) < 4:
        dflist[i] ="000" + dflist[i]
    elif len(dflist[i]) < 5:
        dflist[i] ="00" + dflist[i]
    elif len(dflist[i]) < 6:
        dflist[i] = "0" + dflist[i]
df_esg["기업코드"] = dflist 
df_esg.tail()
```

```python
# df.to_csv(index=False)
df_esg.to_csv('df_esg.csv', index=False)
```
