```python
def screen_name(df):
    df = df[[df.columns[0],'Unnamed: 2','Unnamed: 4','Unnamed: 6']]
    df.columns = df.loc[0]
    df = df.drop([0,1])
    df["등급결정일"] = df["등급결정일"].apply(lambda x: int(str(x).replace(".","")))
    df["현재"][df["현재"] == "취소"] = np.nan
    df["기업명"] = df["기업명"].apply(lambda x: str(x).replace("(주)",""))
    df = df.dropna()
    return df
```

```python
df_n2011 = pd.read_excel("2011-nice.xls")
df_n2011 = screen_name(df_n2011)
df_n2012 = pd.read_excel("2012-nice.xls")
df_n2012 = screen_name(df_n2012)
df_n2012.tail()
```
